```python
a=iris[iris['Species']=='Iris-versicolor']
b=iris[iris['Species']=='Iris-virginica']
c=iris[iris['Species']=='Iris-setosa']

a
```

```python
#Mean, Varience, 

print('Means: ')
print(np.mean(a['PetalLengthCm'])) #Versicolor k petal length ka mean kya hai?
print(np.mean(b['PetalLengthCm']))
print(np.mean(c['PetalLengthCm']))

print('\nStandard Deviation: ')
print(np.std(a['PetalLengthCm'])) #Versicolor k petal length ka standard deviaton kya hai?
print(np.std(b['PetalLengthCm']))
print(np.std(c['PetalLengthCm']))
```

```python
print('Means: ')
print(np.mean(a['PetalWidthCm']))
print(np.mean(b['PetalWidthCm']))
print(np.mean(c['PetalWidthCm']))

print('\nStandard Deviation: ')
print(np.std(a['PetalWidthCm']))
print(np.std(b['PetalWidthCm']))
print(np.std(c['PetalWidthCm']))
```
